```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
from transformers import BitsAndBytesConfig
```

```python
model_name: str = 'microsoft/phi-2'
```

```python
# For the case that gpu not strong enough
nf4_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type='nf4',
    bnb_4bit_compute_dtype=torch.bfloat16
)
```

```python
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    #quantization_config=nf4_config,
    low_cpu_mem_usage=True
)
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
max_new_tokens = 1024
```

```python
model_pipeline = pipeline(
    'text-generation',
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=max_new_tokens,
    device=0,
    pad_token_id=tokenizer.eos_token_id
)
"""
pad_token_id=tokenizer.eos_token_id
  pad_token_id: Specifies the token ID used to pad input sequences to the desired length. Padding ensures that input lengths are consistent for batch processing.
  tokenizer.eos_token_id: The ID of the end-of-sequence (EOS) token, which signals the model to stop generating. This is used here as the padding token for consistency.
"""
```

```python
# For use
from langchain.llms import HuggingFacePipeline

gen_kwargs = {
    'temperature': 0.5
}

llm = HuggingFacePipeline(
    pipeline=model_pipeline,
    model_kwargs=gen_kwargs
)
```

```python
print(llm.invoke('answer, 1 + 1 = ?'))
```
